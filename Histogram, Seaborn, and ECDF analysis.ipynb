```python
x = np.sort(data.dem_share)
y = np.arange(1, len(x) + 1) / len(x)
```

```python
plt.figure()
plt.plot(x, y, marker = '.', linestyle = None)
plt.margins(0.02)
plt.show()
```

```python
plt.figure(figsize=(8,8))
X1 = np.sort(data[data.state == "PA"].dem_share)
Y1 = np.arange(1, len(X1) + 1) / len(X1)
X2 = np.sort(data[data.state == "OH"].dem_share)
Y2 = np.arange(1, len(X2) + 1) / len(X2)
X3 = np.sort(data[data.state == "FL"].dem_share)
Y3 = np.arange(1, len(X3) + 1) / len(X3)
plt.plot(X1, Y1, marker = '.', linestyle = None)
plt.plot(X2, Y2, marker = '.', linestyle = None)
plt.plot(X3, Y3, marker = '.', linestyle = None)
plt.margins(0.02)
plt.show()
```

```python
plt.figure(figsize=(8,8))
X1 = np.sort(data_all[data_all.east_west == "east"].dem_share)
Y1 = np.arange(1, len(X1) + 1) / len(X1)
X2 = np.sort(data_all[data_all.east_west == "west"].dem_share)
Y2 = np.arange(1, len(X2) + 1) / len(X2)
plt.plot(X1, Y1, marker = '.', linestyle = None)
plt.plot(X2, Y2, marker = '.', linestyle = None)
plt.margins(0.02)
plt.xlabel("East and West dem shares distribution")
plt.ylabel("ECDF")
plt.show()
```

```python
Hence, ECDF can be good choice if you want to see how your data is distributed
```
