```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
birthWeight = pd.read_csv('C:\\Users\\deepak\\Downloads\\birthweight_reduced.csv')
```

```python
birthWeight.describe()
```

```python
birthWeight.shape
```

```python
corrBirthWeight = birthWeight.corr()
```

```python
corrBirthWeight #Correlation matrix
```

```python
colormap = sns.diverging_palette(240, 10, n=16)
plt.subplots(figsize=(8,8))
sns.heatmap(corrBirthWeight, cmap=colormap, linewidths=.5) #Adding line width to have space b/w the cells
plt.show()
```

```python
colormap = sns.diverging_palette(240, 10, n=16)
plt.subplots(figsize=(15,15))
mask = np.tri(corrBirthWeight.shape[0], k=0) #Masking the lower triangle
sns.heatmap(corrBirthWeight, cmap=colormap, annot=True, linewidths=.5, mask=mask) 
plt.show()
```

```python
dropSelf = np.zeros_like(corrBirthWeight)
dropSelf[np.triu_indices_from(dropSelf)] = True #Masking self values and upper triangle
colormap = sns.diverging_palette(240, 10, n=16) #16 unique values, therefore need 16 distinct color codes
plt.subplots(figsize=(8,8))
sns.heatmap(corrBirthWeight, cmap=colormap, linewidths=.5, mask=dropSelf)
plt.show()
```
