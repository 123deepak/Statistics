**Find the probability of exactly 4 accidents in a month at a specific part of a roadway, given that average is 5 accidents per month.**

Here Lambda = 5 and we need to compute P(X=4)

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lambda_var = 5
x_var = 4
print(f"Probability of seeing 4 accidents in a month is: {round(poisson.pmf(x_var, lambda_var), 3)}")
```

```python
x_values = np.arange(1, 20, 1)

# Calculate probabilities and store them in a list
pois_probabilities = [round(poisson.pmf(x, lambda_var), 3) for x in x_values]
```

```python
plt.figure(figsize=(8, 3))

# Create the first subplot for the stem plot
plt.stem(x_values, pois_probabilities, markerfmt='bo', linefmt='-b', basefmt=' ')
plt.xlabel("No. of accidents")
plt.ylabel("Probability")
```

```python
#Cummulative
```

```python
x_values = np.arange(1, 20, 1)

# Calculate probabilities and store them in a list
cm_pois_probabilities = [round(poisson.cdf(x, lambda_var), 3) for x in x_values]
```

```python
plt.figure(figsize=(8, 3))

sns.lineplot(x=x_values, y=cm_pois_probabilities, color="b", marker="o")
plt.xlabel("Number of accidents")
plt.ylabel("Cummulative Probability")
```
