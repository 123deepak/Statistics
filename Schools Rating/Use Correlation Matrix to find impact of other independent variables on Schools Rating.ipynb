```python
colormap = sns.diverging_palette(240, 10, n=13)
plt.subplots(figsize=(15,15))
mask = np.tri(corrMatrix.shape[0], k=0) #Masking the lower triangle
sns.heatmap(corrMatrix, cmap=colormap, annot=True, linewidths=.5, mask=mask) 
plt.show()
```
